***2. Initiate	the	spark	context***

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

spark = SparkSession\
        .builder\
        .master('local[*]')\
        .getOrCreate()
sc = spark.sparkContext
```

***Read	Shakespeare	text	from	Google	drive.***

```python
lines = sc.textFile('/content/drive/My Drive/Colab Notebooks/BigDataColabHimaniParikh/Data/words.txt')
```

```python
lines.count()
```

***Split	each	line	into	words.***

```python
words = lines.flatMap(lambda line : line.split(" "))
```

***Assign initial count value to each word.***

```python
tuples = words.map(lambda word : (word,1))
```

***Sum all word count values***

```python
counts = tuples.reduceByKey(lambda a,b: (a+b))
```

***Prepare	the	output	directory.***

```python
!rm -r 'SparkOutPut'
```

***Write word counts to text file in HDFS***

```python
counts.coalesce(1).saveAsTextFile('/content/drive/My Drive/Colab Notebooks/BigDataColabHimaniParikh/SparkOutput')
```

***Question 1:***  If we remove all missing values from the data, how many air pressure at 9am

***Question 2***. If we impute the missing values with the minimum value, how many air
temperature at 9am measurements are less than 42.292?

A. 28

B. 23

C. 1

D. 5

***Answer: A - 28***

```python
from pyspark.sql.functions import min
imputeDf = df
```

```python
for x in imputeDf.columns:
  minValue = removeAllDF.agg(min(x)).first()[0]
  print(x, minValue)
  imputeDf =  imputeDf.na.fill(minValue, [x])
```

```python
result1 = imputeDf[(imputeDf['air_temp_9am'] < 42.292)]
```

```python
result1.count()
```

***Question: 4***  Which column in the weather dataset has the most number of missing values?
